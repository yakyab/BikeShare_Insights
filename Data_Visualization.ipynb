# BikeShare Insights: Data Visualization

This notebook will visualize the insights obtained from the Bike Sharing dataset. We will use Matplotlib for creating the plots.

```python
import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt

# Establish a connection to the SQL database
engine = create_engine('sqlite:///bikeshare.db')

# Load the results from the previous notebook
df_hourly_avg = pd.read_sql('SELECT hr, AVG(cnt) as avg_rentals FROM bike_rentals GROUP BY hr ORDER BY avg_rentals DESC', engine)
df_weather_avg = pd.read_sql('SELECT weathersit, AVG(cnt) as avg_rentals FROM bike_rentals GROUP BY weathersit ORDER BY avg_rentals DESC', engine)
df_weekday_avg = pd.read_sql('SELECT weekday, AVG(cnt) as avg_rentals FROM bike_rentals GROUP BY weekday ORDER BY avg_rentals DESC', engine)
df_season_avg = pd.read_sql('SELECT season, AVG(cnt) as avg_rentals FROM bike_rentals GROUP BY season ORDER BY avg_rentals DESC', engine)
```

Now that we have loaded the data, let's start creating plots to visualize our insights. We'll create bar plots for each of our data frames to better understand the trends in bike rentals.

```python
# Plot the average bike rentals for each hour
plt.figure(figsize=(10,6))
plt.bar(df_hourly_avg['hr'], df_hourly_avg['avg_rentals'])
plt.xlabel('Hour')
plt.ylabel('Average Bike Rentals')
plt.title('Average Bike Rentals per Hour')
plt.show()
```

The above bar plot shows the average number of bike rentals for each hour of the day. This gives us a visual representation of the peak and off-peak hours for bike rentals.

Let's now visualize how different weather conditions affect bike rentals.

```python
# Plot the average bike rentals for each weather situation
plt.figure(figsize=(10,6))
plt.bar(df_weather_avg['weathersit'], df_weather_avg['avg_rentals'])
plt.xlabel('Weather Situation')
plt.ylabel('Average Bike Rentals')
plt.title('Average Bike Rentals per Weather Situation')
plt.show()
```

The above bar plot shows the average number of bike rentals for each weather situation. It visually represents how different weather conditions affect bike rentals.

Finally, let's visualize the variations in bike rental patterns across weekdays and seasons.

```python
# Plot the average bike rentals for each weekday
plt.figure(figsize=(10,6))
plt.bar(df_weekday_avg['weekday'], df_weekday_avg['avg_rentals'])
plt.xlabel('Weekday')
plt.ylabel('Average Bike Rentals')
plt.title('Average Bike Rentals per Weekday')
plt.show()

# Plot the average bike rentals for each season
plt.figure(figsize=(10,6))
plt.bar(df_season_avg['season'], df_season_avg['avg_rentals'])
plt.xlabel('Season')
plt.ylabel('Average Bike Rentals')
plt.title('Average Bike Rentals per Season')
plt.show()
```

The above bar plots show the average number of bike rentals for each weekday and for each season. They visually represent how bike rentals vary depending on the day of the week and the season.
